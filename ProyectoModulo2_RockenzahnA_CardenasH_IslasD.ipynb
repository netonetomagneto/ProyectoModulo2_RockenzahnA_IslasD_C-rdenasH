### **FRACTALES ALEATORIOS PARA LA SIMULACIÓN DE UN PAISAJE TRIDIMENSIONAL** 

### INTRODUCCIÓN 

Imagina que ves una figura y al acercarte ves que esta compuesta de la misma figura, y por más cerca que veas, el patrón se repite una y otra vez; esto es un fragtal, una estructura compleja, repetitiva que a través de la similitud, belleza matemática y se forman de forma natural en varias ocasione.

Entonces, un fractal es una figura, que puede ser espacial o plana, formada por componentes infinitos (Porto & Gardey, 2023). Su principal característica es que su apariencia y la manera en que se distribuye estadísticamente no varía aun cuando se modifique la escala empleada en la observación(Cscazorla, 2012b). 

Un claro ejemplo es el helecho de Barnsley es un fractal autosemejante que lleva el nombre de Michael Barnsley, el matemático que lo describió. Este fractal es famoso por su sorprendente parecido a un helecho real y se utiliza frecuentemente como un ejemplo de cómo las matemáticas pueden imitar la naturaleza. Se genera mediante un sistema de funciones iteradas (IFS), aplicando un conjunto de transformaciones afines que, repetidas a lo largo de muchas iteraciones, producen la estructura característica del helecho. (Wikipedia contributors, 2024)

El nombre de los fractales no fue definido sino hasta 1975 por el matemático Benoît Mandelbrot(Cscazorla, 2012b). El cual definió a estos como "objetos cuya estructura se repite a diferentes escalas"(Cscazorla, 2012b). 

### Propiedades

**Auto-similitud:** "Los fractales tienen patrones que se repiten a diferentes escalas, lo que significa que al observarlos de cerca o de lejos, siguen manteniendo su forma característica." (Barnsley, 1991)

**Complejidad infinita:** "Los fractales tienen un nivel de detalle teóricamente infinito, de modo que siempre que los acerques un poco más, verás nuevos detalles." (Falconer, 2003)

**Generación mediante fórmulas matemáticas:** "Muchos fractales, como el famoso conjunto de Mandelbrot, se crean mediante fórmulas matemáticas iterativas simples, que se aplican repetidamente para formar estructuras complejas." (Peitgen et al., 2006)

Tanto para graficar fractales que se asemejen al helecho de Barnsley como al relieve de nuestro proyecto, se utilizó el principio de escoger un punto (o una coordenada) en la región o espacio y modificarlo por medio de alguna transformación para obtener una nueva coordenada. Este proceso se repitió las veces necesarias hasta obtener todos los puntos requeridos. 

Cuando hablamos de transformaciones nos referimos a los dos posibles casos en los que una función puede alterar algún espacio vectorial, forma, coordenada etc... de alguna de las siguientes maneras: 
- Cuando se habla de transformaciones lineales, se trata de aquellas funciones que pueden hacer rotar o reflejarse sobre su eje a algún objeto sin alterar su forma geométrica. Por ejemplo, este tipo de transformaciones se utiliza mucho en aplicaciones como reducir el tamaño de una imágen.
- En cambio cuando se habla de transformaciones afines, ya no se trata sólo de cambiar la orientación de un ojeto sino que además incluye una traslación del mismo en el espacio. Ya no se respeta el origen, y el objeto puede pasar a ser modificado en su forma, tamaño, posición etc... hasta cambiar radicalmente. 

Las transformaciones afines utilizadas no fueron creadas por nosotros, sino que se obtuvieron las tablas de las transformaciones junto con el modelo general de las mismas. 
Para el primer acercamiento a la graficación de un fractal, se utilizaron los datos conocidos del helecho de Barnsley que se presentan a continuación:

$$\left[\begin{array}{c}x_{k+1}\\ y_{k+1}\end{array}\right]=\left[\begin{array}{cc}a_i & b_i \\ c_i & d_i\end{array}\right]\left[\begin{array}{c}x_k\\ y_k\end{array}\right]+\left[\begin{array}{c}e_i\\f_i\end{array}\right],$$

La manera en la que se interpretan los datos anteriores es que habrá un total de 4 funciones afines que tendrán un impacto en las coordenadas. 
Al seguir la estructura dada por la forma matricial de las transformaciones se tiene que para las primeras coordenadas obtenidas con la función 0, se aplicarían las siguientes ecuaciones: 
$$ x = 0*x + 0*y $$ 
$$ y = 0*x + 0.16*y $$ 
Por lo tanto se observa que esta transformación sólo modifica la parte "y" de la coordenada, haciendo que el punto se mueva hacia arriba en el eje x=0. Debido a esto, se dice que esta función crea "el tallo de la hoja", ya que los puntos se van acomodando en una línea recta en el centro.   
La función 1 mueve el punto un poquito hacia arriba y lo rota hacia la derecha.   
La segunda función rota todos los puntos hacia la izquierda.   
Por último, la tercera función hace que se voltee el punto en su propio eje y lo rota a la derecha.  
Para realizar el proceso completo, se escoge por medio de las probabilidades dadas en la tabla, una de las 4 transformaciones disponibles para modificar la siguiente coordenada y se van graficando los puntos obtenidos. Al final, entre más puntos se grafiquen más marcada quedará la silueta del helecho, y es importante notar que a pesar de que las coordenadas obtenidas puedan parecer aleatorias, en realidad desde el momento en el que se especifican los valores de la tabla, la posición y forma del fractal final ya está determinado. Se puede decir que la posición de los puntos que se grafican es "aleatoria sólo en un cierto rango que ya está delimitado". Esto se puede observar en el hecho de que si se grafican pocos puntos, el acomodo puede parecer completamente aleatorio, pero entre más puntos aparecen se puede ver que la forma de la hoja siempre será igual.  
Hay ejemplos de fractales (como uno que se explicará más adelante) en donde las formas obtenidas son diferentes en cada simulación, sin embargo en el caso del helecho de Barnsley como en el realizado para nuestro proyecto, las formas ya estaban determinadas.

## MODELO QUE REPRESENTA EL PROBLEMA 

Las tablas y modelo obtenidos para nuestro gráfico tridimensional del relieve fueron las siguientes: 

$$\left[\begin{array}{c}x_{k+1}\\ y_{k+1}\\z_{k+1}\end{array}\right]= R_z(\gamma) R_y(\beta) R_x(\alpha) \left[\begin{array}{ccc}a_i & b_i &c_i \\ d_i & e_i & f_i \\ g_i &h_i & j_i\end{array}\right]\left[\begin{array}{c}x_k\\ y_k \\ z_k \end{array}\right]+\left[\begin{array}{c}l_i\\m_i \\ n_i\end{array}\right]$$
$$ R_x(\alpha) \left[\begin{array}{c} 1 & 0 & 0 \\ 0 & cos\alpha & -sin\alpha \\ 0 & sin\alpha & cos\alpha\end{array}\right]$$
$$ R_y(\beta) \left[\begin{array}{c} cos\beta & 0 & sin\beta \\ 0 & 1 & 0 \\ -sin\beta & 0 & cos\beta\end{array}\right]$$
$$ R_z(\gamma) \left[\begin{array}{c} cos\gamma & -sin\gamma & 0 \\ sin\gamma & cos\gamma & 0 \\ 0 & 0 & 1\end{array}\right]$$

(Modificación de un fractal Koch Snowflake, 2024)

```python
import pandas as pd
import numpy as np
table = pd.DataFrame(columns=["i", "A_i", "alpha_i", "beta_i", "gamma_i", "b_i", "p_i"])
table["A_i"] = table["A_i"].astype('object')
table["b_i"] = table["b_i"].astype('object')
table["i"] = [0, 1, 2, 3]
for i in range(4) : table.at[i, "A_i"] = np.eye(3)*0.5
table["alpha_i"] = [0, 0, 0, 2*np.pi/3]
table["beta_i"] = [0, 0, 0, 0]
table["gamma_i"] = [0, 2*np.pi/3, 4*np.pi/3, 0]
b_i = [[[0.5], [0], [0]], [[-0.5], [0], [0]], [[0], [0.5], [0]], [[0], [-0.5], [0]]]
for i in range(4) : table.at[i, "b_i"] = b_i[i]
table["p_i"] = [0.25, 0.25, 0.25, 0.25]
table 
```

## SOLUCIÓN DEL PROBLEMA DE OPTIMIZACIÓN 

El código realizado se muestra a continuación y sigue la misma estructura que el del helecho de Barnsley, sólo que en lugar de interpretar cada transformación en forma de ecuación se decidió definir los datos como matrices para facilitar el cálculo: 

```python
def A_i(): 
    return np.eye(3)*0.5 

def b_i(i): 
    b_i = np.array([[
            [0.5], [0], [0]], 
           [[-0.5], [0], [0]], 
           [[0], [0.5], [0]], 
           [[0], [-0.5], [0]
           ]])  
    return np.reshape(b_i[i],3) 
```

```python
def Rx(alfa): 
    return np.array([(    #np.array pq era un arreglo normal de python pero necesitaba que fuera de numpy y pq se va a multiplicar. No puedes multiplicar los arreglos nornales de python 
        [1,0,0],
        [0, np.cos(alfa), -np.sin(alfa)],
        [0, np.sin(alfa), np.cos(alfa)]
    )])

def Ry(beta): 
    return np.array([(     
        [np.cos(beta),0,np.sin(beta)],
        [0, 1, 0],
        [-np.sin(beta), 0, np.cos(beta)]
    )])

def Rz(gamma): 
    return np.array([(     
        [np.cos(gamma),-np.sin(gamma),0],
        [np.sin(gamma), np.cos(gamma), 0],
        [0, 0, 1]
    )])
def vec_coordenadas(x,y,z):
    return np.array([x,
                     y,
                     z]) 

def fun0(x,y,z):
    coordenadas =  np.dot((Rz(0) @ Ry(0) @ Rx(0)),(np.dot(A_i(),vec_coordenadas(x,y,z)))) + b_i(0)
    return coordenadas #tiene que regresar vector con coordenadas 

def fun1(x,y,z):
    coordenadas =  np.dot((Rz(2.094395) @ Ry(0) @ Rx(0)),(np.dot(A_i(),vec_coordenadas(x,y,z)))) + b_i(1)
    return coordenadas 

def fun2(x,y,z):
    coordenadas =  np.dot((Rz(4.188790) @ Ry(0) @ Rx(0)),(np.dot(A_i(),vec_coordenadas(x,y,z)))) + b_i(2)
    return coordenadas  

def fun3(x,y,z):
    coordenadas =  np.dot((Rz(0) @ Ry(0) @ Rx(2.094395)),(np.dot(A_i(),vec_coordenadas(x,y,z)))) + b_i(3)
    return coordenadas 
```

```python
def simulaciones(n):
    x,y,z = 0,0,0
    lista_x = []
    lista_y= []
    lista_z = [] 
    funciones = [fun0, fun1, fun2, fun3] 
    for j in range(n):
        transformacion = np.random.choice(funciones, p=[0.25, 0.25, 0.25, 0.25])
        respuesta = transformacion(x,y,z)
        respuesta = np.reshape(respuesta, 3)
        x = respuesta[0]
        y = respuesta[1]
        z = respuesta[2]
    
        lista_x.append(x)
        lista_y.append(y)
        lista_z.append(z) 
    arreglo_x = np.array(lista_x)
    arreglo_y = np.array(lista_y)
    arreglo_z = np.array(lista_z)
    return arreglo_x, arreglo_y, arreglo_z
```

```python
def simulaciones2(n):
    x,y,z = 0,0,0
    lista_x = []
    lista_y= []
    lista_z = [] 
    funciones = [fun0_2, fun1_2, fun2_2, fun3_2] 
    for j in range(n):
        transformacion = np.random.choice(funciones, p=[0.25, 0.25, 0.25, 0.25])
        respuesta = transformacion(x,y,z)
        respuesta = np.reshape(respuesta, 3)
        x = respuesta[0]
        y = respuesta[1]
        z = respuesta[2]
    
        lista_x.append(x)
        lista_y.append(y)
        lista_z.append(z) 
    arreglo_x = np.array(lista_x)
    arreglo_y = np.array(lista_y)
    arreglo_z = np.array(lista_z)
    return arreglo_x, arreglo_y, arreglo_z
```

```python
#CÓDIGO NO NUESTRO, TOMADO DE REFERENCIA PARA VER CÓMO SE VERÍA UNA GRÁFICA CON EL MÉTODO DEL MIDPOINT DISPLACEMENT ALGORITHM!!!!


import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def diamond_square(size, scale): 
    grid = np.zeros((size, size))
    grid[0, 0] = np.random.rand() * scale
    grid[0, size-1] = np.random.rand() * scale
    grid[size-1, 0] = np.random.rand() * scale
    grid[size-1, size-1] = np.random.rand() * scale
    
    step_size = size - 1
    
    while step_size > 1:
        half_step = step_size // 2
        
        for x in range(0, size-1, step_size):
            for y in range(0, size-1, step_size):
                avg = (grid[x, y] + grid[x + step_size, y] + grid[x, y + step_size] + grid[x + step_size, y + step_size]) / 4.0
                grid[x + half_step, y + half_step] = avg + (np.random.rand() - 0.5) * scale
        
        for x in range(0, size-1, half_step):
            for y in range((x + half_step) % step_size, size-1, step_size):
                avg = (grid[(x - half_step) % (size-1), y] +
                       grid[(x + half_step) % (size-1), y] +
                       grid[x, (y + half_step) % (size-1)] +
                       grid[x, (y - half_step) % (size-1)]) / 4.0
                grid[x, y] = avg + (np.random.rand() - 0.5) * scale
        
        step_size //= 2
        scale /= 2.0
    
    return grid

def plot_3d_landscape(grid):
    size = grid.shape[0]
    x = np.linspace(0, size - 1, size)
    y = np.linspace(0, size - 1, size)
    x, y = np.meshgrid(x, y)
    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, grid, cmap='terrain')
    
    plt.show()

def monte_carlo_volume(grid, num_samples=100000):
    size = grid.shape[0]
    max_height = np.max(grid)
    
    count_under_surface = 0
    
    for _ in range(num_samples):
        x_rand = np.random.randint(0, size)
        y_rand = np.random.randint(0, size)
        z_rand = np.random.uniform(0, max_height)
        
        if z_rand <= grid[x_rand, y_rand]:
            count_under_surface += 1
    
    volume_fraction = count_under_surface / num_samples
    total_volume = volume_fraction * (size * size * max_height)
    
    return total_volume

# Parámetros
size = 65  # Tamaño de la cuadrícula (debe ser 2^n + 1)
scale = 10.0  # Escala inicial

# Generar el paisaje fractal una vez
landscape = diamond_square(size, scale)

# Graficar el paisaje fractal resultante
plot_3d_landscape(landscape)
```

# GRID 5X5 Midpoint Displacement

```python
def plantilla(tam,n): 
    tamaño = tam**2 +1
    lim = tamaño-1
    grid = np.zeros((tamaño, tamaño))
    grid[0, 0] = np.random.choice(range(n)) 
    grid[0, tamaño-1] = np.random.choice(range(n) )
    grid[tamaño-1, 0] = np.random.choice(range(n) )
    grid[tamaño-1, tamaño-1] = np.random.choice(range(n)) 

    uno = grid[0,0]
    mitad = tamaño //2 #numeros enteros only para coordenadas
    promedio_top = (grid[0, 0] +  grid[lim, 0] )/2 + np.random.rand(1)
    promedio_bot = (grid[0, lim] +  grid[lim, lim] )/2 + np.random.rand(1)
    promedio_left = (grid[0, 0] +  grid[0, lim] )/2 + np.random.rand(1)
    promedio_right = (grid[lim, 0] +  grid[lim, lim] )/2 + np.random.rand(1)
    grid[mitad, 0] = promedio_top[0]
    grid[0,mitad] = promedio_left[0]
    grid[lim,mitad] = promedio_right[0]
    grid[mitad,lim] = promedio_bot[0]
    promedio_cent = (promedio_top + promedio_bot + promedio_left + promedio_right)/4 + np.random.rand(1)
    grid [mitad,mitad] = promedio_cent[0]
    n1_0, n0_1, n1_1, Pn2_1, Pn1_2 = grid3x3(grid[0,0],promedio_top[0], promedio_left[0],promedio_cent[0])
    n3_0, En2_1, n3_1, n4_1, Pn3_2 = grid3x3(grid[mitad, 0],promedio_cent[0], grid[tamaño-1, 0], grid[lim,mitad])
    En1_2, n0_3, n1_3, Pn2_3, n1_4 = grid3x3(grid[0, mitad],grid[0, tamaño-1], promedio_cent[0], promedio_bot[0])
    En3_2, En2_3, n3_3, n4_3, n3_4 = grid3x3(promedio_cent[0],promedio_bot[0], promedio_right[0],grid[tamaño-1, tamaño-1] )
    grid[1,0],grid[0,1],grid[1,1],grid[2,1],grid[1,2] = n1_0, n0_1, n1_1, ((Pn2_1+En2_1)/2), ((Pn1_2+En2_1)/2)
    grid[3,0],grid[3,1],grid[4,1],grid[3,2] = n3_0, n3_1, n4_1, ((Pn3_2+En3_2)/2)
    grid[0,3],grid[1,3],grid[2,3],grid[1,4] = n0_3, n1_3, ((Pn2_3+En2_3)/2),n1_4
    grid[3,3],grid[4,3],grid[3,4] = n3_3,n4_3,n3_4
    return grid
def grid3x3(a,b,c,d):
    tamaño = 3
    lim = tamaño -1
    grid = np.zeros((tamaño,tamaño))
    grid[0, 0] = a 
    grid[0, tamaño-1] = b
    grid[tamaño-1, 0] = c
    grid[tamaño-1, tamaño-1] = d 
    mitad = tamaño //2
    promedio_top = (grid[0, 0] +  grid[lim, 0] )/2 + np.random.rand(1)
    promedio_bot = (grid[0, lim] +  grid[lim, lim] )/2 + np.random.rand(1)
    promedio_left = (grid[0, 0] +  grid[0, lim] )/2 + np.random.rand(1)
    promedio_right = (grid[lim, 0] +  grid[lim, lim] )/2 + np.random.rand(1)
    grid[mitad, 0] = promedio_top[0]
    grid[0,mitad] = promedio_left[0]
    grid[lim,mitad] = promedio_right[0]
    grid[mitad,lim] = promedio_bot[0]
    promedio_cent = (promedio_top + promedio_bot + promedio_left + promedio_right)/4 + np.random.rand(1)
    grid [mitad,mitad] = promedio_cent[0]
    return promedio_top[0],promedio_left[0],promedio_cent[0],promedio_right[0],promedio_bot[0]
```
